```r
library(foreign)
mydata <- read.spss("mydata.sav", header=TRUE, skip=1)
```

```r
str(mydata)
mydata
```

```r
library(rsm)
rsm.y.SOx12 <- rsm(y ~ SO(x1, x2), data = mydata)  # fit second-order (SO) model
```

```r
summary(rsm.y.SOx12)    # show the summary
```

```r
# plot diagnistics
par(mfrow=c(2,4))
```

```r
plot(mydata$x1, rsm.y.SOx12$studres, main="Residuals vs x1")
# horizontal line at zero
abline(h = 0, col = "gray75")
```

```r
# second-order model
contour(rsm.y.SOx12, ~ x1 + x2, image = TRUE, main="second-order model")
persp(rsm.y.SOx12, x2 ~ x1, zlab = "y", main="second-order model")
```
